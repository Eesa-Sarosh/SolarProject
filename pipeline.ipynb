```python
import numpy as np
from keras.models import load_model
from keras.preprocessing.image import load_img, img_to_array
import matplotlib.pyplot as plt
```

```python
model = load_model('solar_farm_model_best.h5')
```

```python
def preprocess_image(image_path, target_size=(150, 150)):
    # Load the image
    img = load_img(image_path, target_size=target_size)
    # Convert the image to a numpy array
    img_array = img_to_array(img)
    # Rescale the image (if the model was trained with rescaled images)
    img_array = img_array / 255.0
    # Expand dimensions to match the model's input shape
    img_array = np.expand_dims(img_array, axis=0)
    return img_array
```

```python
def predict_image(image_path):
    # Preprocess the image
    preprocessed_image = preprocess_image(image_path)
    # Make a prediction
    prediction = model.predict(preprocessed_image)[0][0]
    # Determine the label based on the prediction
    label = 'Solar Farm' if prediction > 0.5 else 'No Solar Farm'
    return prediction, label
```

```python
image_path = '/Users/macbookpro/Desktop/SolarProject/Non_SolarFarms/T10SFJ__-121.11480419982585_38.89536323549093__0.jpg'
prediction, label = predict_image(image_path)

# Print the results
print(f'Prediction Score: {prediction}')
print(f'Predicted Label: {label}')

# Display the image
img = load_img(image_path)
plt.imshow(img)
plt.title(f'Predicted: {label}')
plt.axis('off')
plt.show()
#/Users/macbookpro/Desktop/SolarProject/SolarFarm_Images/T10SGF__-119_augmented_0.jpg
# /Users/macbookpro/Desktop/SolarProject/Non_SolarFarms/T11SLT__-118.3130514710994_33.856094999999996__0.jpg
#/Users/macbookpro/Desktop/SolarProject/Non_SolarFarms/T11SLT__-118.50864618826536_34.23357815899886__0.jpg'
#'/Users/macbookpro/Desktop/SolarProject/SolarFarm_Images/T11SKV__-120.30888141125048_35.84303426435679__0.jpg'
#'/Users/macbookpro/Desktop/SolarProject/SolarFarm_Images/T10SGF__-120.32292260394885_36.53733458654228__0.jpg'
#/Users/macbookpro/Desktop/SolarProject/Non_SolarFarms/image__98537.1661151144.png
#'/Users/macbookpro/Desktop/SolarProject/SolarFarm_Images/T11SKA__-119.92983623209042_36.244404777839186__1.jpg'
#'/Users/macbookpro/Desktop/SolarProject/SolarFarm_Images/T04QEJ__-157.9972097161163_21.46796432372588__0.jpg
#/Users/macbookpro/Desktop/SolarProject/Non_SolarFarms/T10SFJ__-121.11480419982585_38.89536323549093__0.jpg'
#'/Users/macbookpro/Desktop/SolarProject/SolarFarm_Images/T11SKA__-119.92983623209042_36.244404777839186__1.jpg'
```
